**Histogram Distribusi Nilai Matematika**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['nilai_mat'], bins=10, color='cyan', edgecolor='black')
plt.title('Distribusi Nilai Matematika', fontsize=14)
plt.xlabel('Nilai Matematika', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Grafik ini menunjukkan bahwa sebagian besar mahasiswa memiliki nilai matematika di rentang 60-80.

**Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas**

```python
faculty_counts = df['fakultas'].value_counts()

plt.figure(figsize=(8, 6))
plt.barh(faculty_counts.index, faculty_counts.values, color='cyan', edgecolor='black')
plt.title('Jumlah Mahasiswa per Fakultas', fontsize=14)
plt.xlabel('Jumlah Mahasiswa', fontsize=12)
plt.ylabel('Fakultas', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

Fakultas dengan jumlah mahasiswa terbanyak adalah FTI

**Diagram pie banyaknya mahasiswa masing-masing fakultas**

```python
plt.figure(figsize=(8, 8))
plt.pie(faculty_counts, labels=faculty_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Proporsi Mahasiswa per Fakultas', fontsize=14)
plt.show()
```

Menampilkan proporsi mahasiswa di tiap fakultas dalam bentuk persentase

**Berdasarkan diagram pie dan batang horizontal manakah fakultas dengan mahasiswa terbanyak?**

*Manakah diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?*

```python
mahasiswa_fakultas = faculty_counts.idxmax()
mahasiswa_fakultas
```

Fakultas dengan mahasiswa terbanyak: Fakultas Teknologi Industri (FTI).

Diagram batang lebih baik dalam menunjukkan jumlah mahasiswa per fakultas secara langsung, karena mempermudah perbandingan visual.

**Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y**

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['nilai_kim'], df['nilai_fis'], alpha=0.7, c='yellow', edgecolor='black')
plt.title('Scatter Plot Nilai Kimia vs. Nilai Fisika', fontsize=14)
plt.xlabel('Nilai Kimia', fontsize=12)
plt.ylabel('Nilai Fisika', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

mahasiswa_fakultas
```

Menunjukkan hubungan positif antara nilai kimia (sumbu x) dan fisika (sumbu y)
