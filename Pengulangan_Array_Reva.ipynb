# Tugas 1.
**Buat flowchart dari contoh-3 dan jelaskan kenapa kasus pada contoh-3 menggunakan while bukannya for**

image.png

Pada contoh-3, loop while digunakan karena kita tidak tahu sebelumnya berapa banyak bilangan integer yang akan dimasukkan oleh pengguna. Pengguna dapat memasukkan sejumlah bilangan integer yang tidak diketahui sebelumnya, dan loop while akan terus berulang sampai pengguna memasukkan bilangan -999 sebagai tanda akhir input.

Jika kita menggunakan loop for, kita harus menentukan jumlah iterasi terlebih dahulu. Dalam kasus ini, kita tidak tahu berapa banyak bilangan integer yang akan dimasukkan, sehingga loop for tidak sesuai.

Loop while lebih fleksibel dalam hal ini karena dapat berulang selama kondisi tertentu terpenuhi, yaitu sampai pengguna memasukkan -999.

# Tugas 2.
**Buatlah algoritma/program yang membaca sebuah nilai integer positif, misalnya N, dan menjumlahkan (serta menampilkan) semua bilangan kelipatan 5 antara 1 s.d. N.**

```python
# Input: Read a positive integer N from the user
N = int(input("Masukkan nilai N: "))

# Initialize an array to store multiples of 5 and a variable to store the sum
multiples = []
total_sum = 0

# Find all multiples of 5 between 1 and N, add them to the array, and update the total sum
for i in range(5, N + 1, 5):
    multiples.append(i)
    total_sum += i

# Output the results
print("Input N:", N)
print("Output:", total_sum)

# Display detailed information based on the presence of multiples
if multiples:
    print("Keterangan: Bilangan kelipatan 5 antara 1 s.d.", N, "adalah", multiples)
else:
    print("Keterangan: Tidak ada bilangan kelipatan 5 antara 1 s.d.", N)
```

# Tugas 3.
**Buatlah 3 versi program yang menerima masukan sebuah integer positif, misalnya N (asumsi N>0), dan menuliskan lirik lagu Anak Ayam di atas dengan menggunakan perulangan for dan while.**

*VERSI 1 Menggunakan for Loop*


```python
N = int(input("Masukkan jumlah anak ayam (N): "))
lyrics = []

for i in range(N, 0, -1):
    if i > 1:
        lyrics.append(f"Anak ayam turunlah {i}")
        lyrics.append(f"Mati satu tinggallah {i - 1}")
    else:
        lyrics.append(f"Anak ayam turunlah 1")
        lyrics.append("Mati satu tinggallah induknya")

# Print lyrics
for line in lyrics:
    print(line)
```

*VERSI 2 Menggunakan while Loop*

```python
N = int(input("Masukkan jumlah anak ayam (N): "))
lyrics = []

while N > 0:
    if N > 1:
        lyrics.append(f"Anak ayam turunlah {N}")
        lyrics.append(f"Mati satu tinggallah {N - 1}")
    else:
        lyrics.append(f"Anak ayam turunlah 1")
        lyrics.append("Mati satu tinggallah induknya")
    N -= 1

# Print lyrics
for line in lyrics:
    print(line)
```

*VERSI 3 Menggunakan for Loop dalam Urutan Terbalik (dengan array)*

```python
N = int(input("Masukkan jumlah anak ayam (N): "))
lyrics = [f"Anak ayam turunlah {i}\nMati satu tinggallah {i - 1}" if i > 1 else "Anak ayam turunlah 1\nMati satu tinggallah induknya" for i in range(N, 0, -1)]

# Print lyrics
for line in lyrics:
    print(line)
```
