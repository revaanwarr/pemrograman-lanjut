Problem (1) Kelipatan 10 terkecil

```python
import math
def kelipatan_10_terkecil(N):
  return int(math.ceil(N/10.0)*10)

print(kelipatan_10_terkecil(5))
print(kelipatan_10_terkecil(1234))
```

(2) Segitiga Angka

```python
def segitiga_angka(N):
  for i in range(1, N+1):
    for j in range(1, i+1):
      print(j, end="")
    print()

segitiga_angka(5)
```

**Loop yang tepat**
Untuk masalah ini, loop bersarang(nested lopp) adalah yang terbaik karena kita harus mencetak angka bertingkat untuk setiap baris

```python
#Test case 1
def siap_bang_jago(N, A, B, C):
 for i in range(1, N+1):
  output = ""
  if i % A == 0:
    output += "Siap"
  if i % B == 0:
    output += "Bang"
  if i % C == 0:
    output += "Jago"
  if output == "":
    output = str(i)
  print (output)
siap_bang_jago(12, 2, 3, 4)
```

**Interpretasi coding:**
Kode ini menjalankan fungsi untuk memeriksa kelipatan dari beberapa angka sekaligus dan mencetak hasil berdasarkan kondisi yang diberikan. Kata-kata dapat muncul bersamaan jika angka adalah kelipatan dari satu variabel, sehingga memungkinkan kombinasi seperti "SiapBang" atau "SiapBangJago"

**Interpretasi hasil:**
Hasil akhir dari setiap iterasi akan dicetak dalam satu baris. Untuk N=12, A=2, B=3, C=4, program akan memeriksa angka 1 hingga 12. Untuk angka 4, karena 4 habis dibagi 2 dan 4, maka hasil yang ditampilkan adalah "SiapJago"
